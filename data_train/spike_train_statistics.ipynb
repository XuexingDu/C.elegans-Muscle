```python
import pandas as pd
import matplotlib.pyplot as plt
# Load the data
file_path_1 = 'spike_train.csv'
file_path_2 = 'Spike_train2.csv'
file_path_3 = 'Spike_train3.csv' 

spike_train_data = pd.read_csv(file_path_1)
spike_train_data_2 = pd.read_csv(file_path_2)
spike_train_data_3 = pd.read_csv(file_path_3)

# Display the first few rows of the dataframe
spike_train_data.head()
```

```python
# Extracting time column
time = spike_train_data.iloc[:, 0]

# Number of spike trains to plot (for demonstration, plotting the first 5 spike trains)
# num_spike_trains_to_plot = min(5, spike_train_data.shape[1] - 1)
num_spike_trains_to_plot = spike_train_data.shape[1] - 1

# Creating subplots
fig, axes = plt.subplots(num_spike_trains_to_plot, 1, figsize=(10, num_spike_trains_to_plot * 3))
fig.tight_layout(pad=3.0)

# Plot each spike train
for i in range(num_spike_trains_to_plot):
    spike_train = spike_train_data.iloc[:, i + 1]
    if num_spike_trains_to_plot == 1:
        axes.plot(time, spike_train)
        axes.set_title(f'Spike Train {i+1}')
        axes.set_xlabel('Time (s)')
        axes.set_ylabel('Amplitude')
    else:
        axes[i].plot(time, spike_train)
        axes[i].set_title(f'Spike Train {i+1}')
        axes[i].set_xlabel('Time (s)')
        axes[i].set_ylabel('Amplitude')

plt.show()
```
